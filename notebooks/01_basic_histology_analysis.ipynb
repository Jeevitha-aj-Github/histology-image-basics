# Basic Histology Image Analysis

Below is your histology image embedded directly so GitHub can display it without running any code:

![Histology Image](../data/histo_01.png)

```python
from PIL import Image
import matplotlib.pyplot as plt

# load original image
img = Image.open('../data/histo_01.png')
plt.imshow(img)
plt.axis('off')
```

```python
# convert to grayscale
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
```

```python
import numpy as np

# grayscale as numpy array
gray_arr = np.array(img_gray)

# simple threshold: darker = tissue
threshold = gray_arr.mean()
mask = gray_arr < threshold

plt.imshow(mask, cmap='gray')
plt.axis('off')
```
